```python
import pandas as pd
import re

df = pd.read_csv('../csv/Final.csv')
df
```

```python
df = df[df['Texto'].str.len() >= 10]
patron = r'\b\w+\b [^\W\d_]+ \w+\b'

# Filtrar filas con texto que cumple el patrón
df_filtrado = df[df['Texto'].str.match(patron, na=False)]
df_filtrado
```

```python
df_sorted = df_filtrado.sort_values(
    by=['Texto', 'Fecha', 'Nombre Usuario', 'Fuente', 'Concepto asociado', 'Género', 'Grupo Étareo', 'Escolaridad'])

# Exportar el DataFrame filtrado a un archivo Excel
df_filtrado.to_excel('df_filtrado.xlsx', index=False)
```
